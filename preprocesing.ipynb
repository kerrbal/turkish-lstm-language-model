```python
import sentencepiece
```

```python
sentencepiece.SentencePieceTrainer.train(input = "source/stories.txt", model_prefix = "omer_s", vocab_size = 16000, model_type = "bpe", 
                                         character_coverage = 1.0)
```

```python
sp = sentencepiece.SentencePieceProcessor(model_file = "omer_s.model")
```

```python
sp.__len__()
```

```python
import numpy as np
```

```python
#read_all_tokens
with open("source/stories.txt", "r", encoding = "utf-8") as f:
    text = f.read()

ids = sp.encode(text, out_type = int)
the_array = np.array(ids, dtype = np.int32)
np.save("train_.npy", the_array)
```
